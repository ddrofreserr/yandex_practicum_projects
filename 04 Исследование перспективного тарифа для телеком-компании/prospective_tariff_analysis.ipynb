Для веб-трафика отдельные сессии не считаются. Вместо этого общая сумма за месяц округляется в бо́льшую сторону. Если абонент использует 1025 мегабайт в этом месяце, с него возьмут плату за 2 гигабайта. Примите, что неиспользованные за предыдущий месяц звонки, смс, интернет на следующий месяц не переносятся.

### Добавление информации по месяцам

Чтобы не загромождать таблицу с информацией о пользователях, сделаем таблицу для информациии по месяцам, так как у нас данные всего за 12 месяцев и о 500 пользователях

```python
# создаем подходящий словарь
data = {'user_id': [id for id in users_data['user_id'] for _ in range(12)],
        'month': [i for i in range(1, 13)] * len(users_data['user_id'])}

# преобразуем в датафрейм
data_by_months = pd.DataFrame(data=data)
# проверка результата
display(data_by_months.head(15))
```

Поскольку записей в каждой таблице очень много, чтобы обработать по отдельности каждую строку, сделаем сводные таблицы по месяцам и юзерам для:
* количества сделанных звонков,
* количества израсходованных минут разговора,
* количества отправленных сообщений,
* объема израсходованного интернет-трафика;

Для этого добавим в каждую таблицу колонку "месяц".

```python
# проходимся циклом по всем таблицам
for table in [calls_data, messages_data, internet_data]:
    date_column_name = list(filter(lambda x: 'date' in x, table.columns))[0]
    table['month'] = table[date_column_name].dt.month
```

```python
# информация для звонков
calls_by_month = calls_data.pivot_table(index=['user_id', 'month'],
                                        values=['duration_round'],
                                        aggfunc=['count', 'sum'])
# делаем столбцы user_id и month из индексов
calls_by_month = calls_by_month.reset_index(1).reset_index(0)
# переименовываем столбцы
calls_by_month.columns = ['user_id', 'month', 'calls_count', 'calls_minutes']

# информация для сообщений
messages_by_month = messages_data.pivot_table(index=['user_id', 'month'],
                                              values=['id'],
                                              aggfunc='count')
# так же преобразуем
messages_by_month = messages_by_month.reset_index(1).reset_index(0)
messages_by_month.columns = ['user_id', 'month', 'messages_count']

# информация для интернета
internet_by_month = internet_data.pivot_table(index=['user_id', 'month'],
                                              values='mb_used',
                                              aggfunc='sum')
# все то же самое
internet_by_month = internet_by_month.reset_index(1).reset_index(0)
internet_by_month.columns = ['user_id', 'month', 'internet_mb']
```

Напишем функцию для добавления в таблицу data_by_month, работающую для каждой pivot_table

```python
# проходимся по каждой таблице
for table in [calls_by_month, messages_by_month, internet_by_month]:
    data_by_months = pd.merge(data_by_months, table, 
                              on=['user_id', 'month'], how='left')

# избавимся от месяцев без какой-либо активности
data_by_months = data_by_months.dropna(how='all', subset=['calls_count', 
                                                          'calls_minutes',	
                                                          'messages_count',	
                                                          'internet_mb'])

display(data_by_months.head(20))
```

Исправим немного таблицу, заменим пропуски на нули и исправим типы данных в столбцах звонков и сообщений на целочисленные

```python
# заполнение пропусков
data_by_months = data_by_months.fillna(0)

# приведение к целочисленному типу данных
data_by_months['calls_count'] = data_by_months['calls_count'].astype('int')
data_by_months['calls_minutes'] = data_by_months['calls_minutes'].astype('int')
data_by_months['messages_count'] = data_by_months['messages_count'].astype('int')
```

Теперь необходимо округлить использованные мегабайты интернета до гигабайт:

```python
# создадим для этого отдельный столбец
data_by_months['internet_gb'] = data_by_months['internet_mb'] / 1024
data_by_months['internet_gb'] = data_by_months['internet_gb'].apply(math.ceil)
```

А также переведем в таблице с информацией о тарифах мегабайты в гигабайты:

```python
tariff_data['gb_per_month_included'] = tariff_data['mb_per_month_included'] // 1024
```

Теперь посчитаем помесячную выручку с каждого пользователя, для этого добавим столбцы с подсчетом минут, сообщений и интернет трафика сверх тарифа, чтобы разделить абонентов по тарифам и для удобства дальнейшего анализа, добавим в data_by_month столбец


```python
# список айди пользователей, использующих тариф смарт
smart_users = list(users_data.query('tariff == "smart"')['user_id'])

# функция для записи
def tariff_checker(user_id):
    if user_id in smart_users:
        return 'smart'
    else:
        return 'ultra'

data_by_months['tariff'] = data_by_months['user_id'].apply(tariff_checker)
```

```python
# функция для подсчета затрат вне тарифа
def price_over_count(row, to_count): # to_count может принимать messages, gb или minutes
    name_of_column = [name for name in row.index if (to_count in name)][0]
    tariff = row['tariff']

    included_col = [col for col in tariff_data.columns if ('included' in col
                                                           ) and (to_count in col)][0]
    
    included = int(tariff_data.query('tariff_name == @tariff')[included_col].iloc[0])

    price_per_col = [col for col in tariff_data.columns if ('per' in col
                                                            ) and (to_count in col)][0]

    price_per = int(tariff_data.query('tariff_name == @tariff')[price_per_col].iloc[0])
    
    result = (row[name_of_column] - included) * price_per
    if result > 0:
        return result
    return 0
```

```python
#подсчет для минут звонков
data_by_months['over_calls_price'] = data_by_months.apply(price_over_count, 
                                                          axis=1,
                                                          args=('minute', ))

#подсчет для количества сообщений
data_by_months['over_messages_price'] = data_by_months.apply(price_over_count, 
                                                          axis=1,
                                                          args=('message', ))

# подсчет для интернет трафика
data_by_months['over_internet_price'] = data_by_months.apply(price_over_count, 
                                                          axis=1,
                                                          args=('gb', ))
```

Теперь добавим общую цену за месяц:

```python
# функция для подсчета
def add_total_price(row):
    tariff = row['tariff']
    price_per_month = int(tariff_data.query('tariff_name == @tariff')['rub_monthly_fee'].iloc[0])
    return (row['over_calls_price'] + 
            row['over_messages_price'] + 
            row['over_internet_price'] +
            price_per_month)

data_by_months['total_price'] = data_by_months.apply(add_total_price, axis=1)
```

## Анализ поведения клиентов

### Трата пакета

Проанализируем отдельно продолжительность звонков, количество сообщений и объём интернет-трафика пользователей каждого тарифа, посчитаем среднее, дисперсию и стандартное отклонение.

```python
# функция для прохода по показателям
def show(column):
    # оценка разброса данных
    for tariff in ['smart', 'ultra']:
        by_tariff = data_by_months.query('tariff == @tariff')
        print(f'Для {column} при тарифе {tariff}:',
            f'Среднее: {by_tariff[column].mean()}', 
            f'Дисперсия: {np.var(by_tariff[column])}',
            f'Стандартное отклонение: {np.sqrt(np.var(by_tariff[column]))}', 
            sep='\n', end='\n\n')
    
    # гистограммы по тарифам
    pe.histogram(data_by_months, 
                 x=column, 
                 color='tariff', 
                 marginal="box",
                 barmode='overlay').show()
```

```python
for col in ['calls_minutes', 'messages_count', 'internet_mb']:
    show(col)
```

Видим гистограммы нормального распределения, при тарифе "Смарт" разброс значений много меньше, чем при тарифе "Ультра" для всего пакета

## Проверка гипотезы

### Средняя выручка пользователей тарифов «Ультра» и «Смарт» различаются

Будем судить о средней выручке по затратам на тариф

$H_{0}$ - средняя выручка пользователей разных тарифов различаются

Проверим гипотезу о равенстве генеральных совокупностей для разных тарифов с помощью t-теста:

```python
# зададим критический уровень
alpha = .05

res = st.ttest_ind(data_by_months.query('tariff == "smart"')['total_price'], 
                   data_by_months.query('tariff == "ultra"')['total_price'], 
                   equal_var=False) 
# принимаем False так как дисперсии сильно различаются

print('p_value:', res.pvalue)

if res.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Действительно, p-value невероятно мало, то есть при правильности нулевой гипотезы, вероятность наблюдать такие значения практически невозможно, отвергаем нулевую гипотезу

### Cредняя выручка пользователей из Москвы отличается от выручки пользователей из других регионов

$H_{0}$ - средняя выручка пользователей не равна

$H_{1}$ - средняя выручка пользователей из Москвы выше, чем у пользователей из других регионов

Проверять гипотезу будем так же через t-теста

```python
# запишем индексы пользователей из Москвы
users_from_msk = users_data.query('city == "Москва"')['user_id']

res = st.ttest_ind(data_by_months.query('user_id in @users_from_msk'
                                        )['total_price'],
                   data_by_months.query('user_id not in @users_from_msk'
                                        )['total_price'],
                  equal_var=False)

print('p_value:', res.pvalue)

# критический уровень оставим прежним
if res.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

P-value намного больше критического уровня статистической значимости, приходим к выводу, что выручка слабо зависит от региона
